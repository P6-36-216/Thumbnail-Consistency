### Resize + save + blur-ridden

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('/home/pramatha/thumb.jpg',cv2.IMREAD_UNCHANGED)

print('Original dimensions:',img.shape)
scale = 400
width = int(img.shape[1]*scale / 100)
height = int(img.shape[0]*scale / 100)
dim = (width,height)

resized = cv2.resize(img,dim,interpolation=cv2.INTER_AREA)
print('Resized dimensions:',resized.shape)
cv2.imwrite('/home/pramatha/image.png',resized)

sharpen_kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
sharpen = cv2.filter2D(resized, -1, sharpen_kernel)

cv2.imshow('sharpen', sharpen)
cv2.waitKey()

dim=(512,640)
image = cv2.imread('/home/pramatha/IMG_20170619_151258.jpg')
res = cv2.resize(image,dim,interpolation=cv2.INTER_AREA)
cv2.imwrite('/home/pramatha/resized.jpg',res)

cv2.imshow('Original',res)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Grayscale analysis / Plotting Histograms

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/home/pramatha/thumb.jpg')
gra_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gra_img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
gra_img_eqhist = cv2.equalizeHist(gra_img)
hist = cv2.calcHist(gra_img_eqhist,[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
hist,bins = np.histogram(img.ravel(),256,[0,256])
plt.plot(hist)
plt.show()
```

```python
img = cv2.imread('/home/pramatha/thumb.jpg',0)
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

```python
img = cv2.imread('/home/pramatha/thumb.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

### Image negatives

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('/home/pramatha/thumb.jpg',0)
inv_img = 255 - img
fig = plt.figure(figsize=(15,5))

ax= fig.add_subplot(121)
plt.title('Original',fontsize=18)
plt.imshow(img,cmap='gray')

ax = fig.add_subplot(122)
plt.imshow(inv_img,cmap='gray')
ax.set_title('Inverse transform',fontsize=18)
```

### Log transformations

### Power law transformations

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('/home/pramatha/thumb.jpg')
im1 = image/255.0
img_pow = cv2.pow(im1,1.2)

scale = 400
width = int(image.shape[1]*scale / 100)
height = int(image.shape[0]*scale / 100)
dim = (width,height)

img = cv2.resize(image,dim,interpolation=cv2.INTER_AREA)

cv2.imshow('Original image',img)
cv2.waitKey()

img_pow = cv2.resize(img_pow,dim,interpolation=cv2.INTER_AREA)
cv2.imwrite('/home/pramatha/power_law.jpg',img_pow)

cv2.imshow('Next image',img_pow)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Blur-Ridden 1

```python
import cv2
import numpy as np
```

```python
image = cv2.imread('/home/pramatha/thumb.jpg')
Gaussian = cv2.GaussianBlur(image,(7,7),0)
Median = cv2.medianBlur(image,5)
Bilateral = cv2.bilateralFilter(image,9,75,75)

scale = 400
width = int(image.shape[1]*scale / 100)
height = int(image.shape[0]*scale / 100)
dim = (width,height)

resized1 = cv2.resize(image,dim,interpolation=cv2.INTER_AREA)
resized2 = cv2.resize(Gaussian,dim,interpolation=cv2.INTER_AREA)
resized3 = cv2.resize(Median,dim,interpolation=cv2.INTER_AREA)
resized4 = cv2.resize(Bilateral,dim,interpolation=cv2.INTER_AREA)

cv2.imshow('Original image',resized1)
cv2.waitKey()

cv2.imshow('Gaussian Blurring',resized2)
cv2.waitKey()

cv2.imshow('Median Blurring',resized3)
cv2.waitKey()

cv2.imshow('Bilateral Blurring',resized4)
cv2.waitKey()
cv2.destroyAllWindows()
```
