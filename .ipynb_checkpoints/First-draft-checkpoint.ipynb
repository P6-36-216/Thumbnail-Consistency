```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image,ImageChops
```

### Extract thumbnail 

```python
def thumbnail():
    try:
        image = Image.open('picture.jpg')
        image.thumbnail((128,160))
        image.save('picture-thumbnail.jpg')
        
        thumb = Image.open('picture-thumbnail.jpg')
        thumb.show()
    except IOError:
        pass
thumbnail()

thumb = cv2.imread('picture-thumbnail.jpg')        
width = thumb.shape[1]
height = thumb.shape[0]
print(width,height)
```

### Scale down Forged image to the size of thumbnail

```python
dim = (width,height)

orig = cv2.imread('picture.jpg')
res = cv2.resize(orig,dim,interpolation=cv2.INTER_AREA)

#print('Dimensions of resized-original-image: ',res.shape[1],res.shape[0])
#print('Dimensions of thumbnail: ',thumb.shape[1],thumb.shape[0])
print(res.shape,thumb.shape)
cv2.imshow('Original Image',res)
cv2.waitKey()

cv2.imshow('Thumbnail',thumb)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Comparison

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([res],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
for i,col in enumerate(color):
    histr = cv2.calcHist([thumb],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

### KDE Algorithm

```python
hist = cv2.calcHist([thumb],[0],None,[256],[0,256])
plt.hist(hist,bins=30,normed=True)
```

### Comparison using MSE

```python
#anot = cv2.imread('')

def mse(im1,im2):
    err = np.sum((im1.astype("float") - im2.astype("float")) ** 2)
    err /= float(im1.shape[0] * im1.shape[1])
    print(err)

mse(res,thumb)
```
